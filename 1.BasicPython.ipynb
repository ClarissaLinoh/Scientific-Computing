```python
#VARIABLE ASSIGMENT

a, b = 8, 2.5 #nilai a menjadi 1 dan nilai b menjadi 2.5
print(a, b) #output 1 2.5
print(a + b) #output 10.5 --> float --> karena operasi antara float + integer = float

#Tipe data secara otomatis berdasarkan value yang di assign
a = True        #Boolean
b = "Bravo"     #String
c = 2.8         #Float
d = 12          #Integer

print(a, b, c, d)
print(f"Nilai a {a}, nilai b {b}, nilai c {c}, nilai d {d}") #String formatting
```

```python
#SELECTION

var1 = 99
var2 = 99.9

# IF ELIF ELSE
if var1 > var2:
    print("Yes")
elif var1 == var2 :
    print("Equal")
else :
    print("No")

# MATCH CASE --> tidak ada di ipynb ??

#TERNARY

x = True if var1 < var2 else False
if x == True:
    print("Yes")
else :
    print("No")
```

```python
# AND & OR --> Logical Operator

if True == True and True == False:  #hasilnya "No" karena True == True benar, tetapi True == False salah (AND harus keduanya benar)
    print("Yes")
else:
    print("No")

if True == True or True == False:  #hasilnya "Yes" karena True == True benar, dan karena ini OR maka salah satu benar bisa
    print("Yes")
else:
    print("No")


# IN & NOT IN
# Searching secara otomatis dengan cara yang paling optimal
# Python a bit slow, tp jika library nya ditulis dengan bahasa low level (cth c) akan menjadi cepat juga

arr = [1, 2, 3, 4, False, 2.5, "Sunday"]
if "Sunday" in arr:
    print("Exist")
else :
    print("Not Exist")

if 8 in arr:
    print("Exist")
else :
    print("Not Exist")
```

```python
#REPETITION
#--> Pengulangan yang dilakukan berdasarkan kondisi tertentu

# FOR
# --> mengulangi blok kode dengan JUMLAH PENGULANGAN DIKETAHUI

# For dalam string
for i in "apple":
    print(i, end=" ")

print("\n")

# For dalam array (list, tuple, set)
arr = [1, 4, 2, 3, 1, 2]
for i in arr:
    print(arr[i], end=" ")

print("\n")

# dengan RANGE
# satu parameter
for i in range(8):      # start dari indeks ke 0 sampai 7 --> iterasi sebanyak 8 kali
    print("*", end=" ")

print("\n")

#dua parameter
for i in range(2, 8):   # start dari indeks ke 2 sampai 7 --> iterasi 8 - 2 = 6 kali
    print("#", end=" ")

print("\n")

#tiga parameter
for i in range(2, 18, 2):  # start dari indeks ke 2 sampai 18 dan ditambah 2 per iterasi (18 - 2) / 2 = 16 / 2 = 8 kali
    print("$", end=" ")



#WHILE
#--> pengulangan yang dilakukan selama kondisi tetap TRUE

i = 0                   # inisiasi nilai variable
while i < 6:
    print(i, end=" ")
    i += 1              # WAJIB ada statemen yang MENGUBAH value variable sehingga MENCAPAI KONDISI --> increment / decrement

print("\n")


# JUMP OPERATOR --> break & continue

#Break 
# -->Jika kondisi terpenuhi iterasi dihentikan

for i in range(2, 100):
    if(i == 4): break   #ketika mencapai i = 4 iterasi dihentikan --> output hanya 2 dan 3
    print(i, end=" ")

print("\n")

x = 10
while x > 0:
    if(x == 6): break   #ketika mencapai x = 6 iterasi dihentikan --> output hanya 10, 9, 8, 7
    print(x, end=" ")
    x-=1

print("\n")


#Continue
#  --> Jika kondisi terpenuhi, SKIP iterasi

for i in range(2, 10):
    if(i == 4):
        i+=1        # HARUS ada PERUBAHAN NILAI --> jika tidak akan stuck di continue terus --> nilai i tetap 4 / infinite loop
        continue    #ketika mencapai i = 4 iterasi diskip --> angka 4 tidak ditampilkan
    print(i, end=" ")

print("\n")

x = 10
while x > 0:
    if(x == 6): 
        x -= 1      # HARUS ada PERUBAHAN NILAI --> jika tidak akan stuck di continue terus --> nilai i tetap 4 / infinite loop
        continue    #ketika mencapai x = 6 iterasi diskip --> angkat 6 tidak ditampilkan
    print(x, end=" ")
    x -= 1

print("\n")


#ENUMERATE ARRAY
# --> Mengambil indeks dan value nya

arr = [True, 2.1, 44, 18888888, False, "Saturday"]
for i, x in enumerate(arr):
    print(i+1, x)
```

```python
# LIST --> Bisa diubah, tidak harus urut, bisa duplikat

list1 = [1, 1, 2, 3, 2, 1, 1]
print(list1)       #Output --> [1, 1, 2, 3, 2, 1, 1]

#CHANGE element
list1[4] = 8
print(list1)

#SORT element
list1.sort
print(list1)

list1.sort(reverse=True) #sortir terbalik
print(list1)

# INSERT / APPEND --> hanya bisa satu persatu element nya
list1.insert(4, 12) #menambahkan 12 di indeks ke 4 --> menggeser yang lain
print(list1)

list1.append(14)    #menambahkan di paling belakang / indeks terakhir
print(list1)



```

```python
# SET --> terurut, tidak ada duplikat, BISA dihapus & ditambahkan elementnya

set1 = {1, 22, 3, 2, 2, 4, 5}
print(set1)        #Output --> {1, 2, 3, 4, 5, 22}

#REMOVE
set1.remove(4)  # angka 4 akan dicari dan dihapus
print(set1)

#UPDATE
set1.update({19})
print(set1)

set1.update({5, 6, 4, 3, 22, 188})  #beberapa element sekaligus
print(set1)
```

```python
# TUPLE --> Tidak bisa diubah, tidak urut
# untuk menyimpan angka-angka seperti konstanta

tuple1 = (3, 3, 2, 1, 3, 8)
print(tuple1)       #Output --> (3, 3, 2, 1, 3, 8)

#bisa di typecasting ke list
tupleToList = list(tuple1)
print(tupleToList)
```
